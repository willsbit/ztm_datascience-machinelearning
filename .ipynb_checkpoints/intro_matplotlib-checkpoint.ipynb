```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

plt.plot(x,y)
plt.show()
```

```python
# 1º método
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2º método
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# 3º método (recomendado)
fig, ax = plt.subplots(figsize=(9, 8))
ax.plot(x, y)

ax.set(title = "Simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")
plt.show()

fig.savefig("./matplotlib_exemplo1.png")
```

## Scatter plot

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

## Bar plot

```python
# make a plot from a dict
nut_butter_prices = {'Almond butter': 10,
                     "Peanut butter": 8,
                     "Cashew butter": 15}

fig, ax = plt.subplots(dpi=100)
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())

ax.set(title = "Butter prices",
       xlabel = "type of butter",
       ylabel = "price ($)")

plt.show()
```

## Histogram

```python
fig, ax = plt.subplots(dpi=100)
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))

ax.set(title = "Butter prices",
       xlabel = "price ($)",
       ylabel = "type of butter")

plt.show()
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

## Duas opções de subplots

#### Opção 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

ax1.plot(x, x/2)
ax2.scatter(x, np.cos(x))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(x)
plt.show()
```

#### Opção 2

```python
fig, ax =  plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

ax[0, 0].plot(x, x/2) ## row, column
ax[0, 1].scatter(x, np.tanh(x))
ax[1, 0].scatter(x, np.sinh(x))
ax[1, 1].scatter(x, np.cosh(x))

plt.show()
```

### Plots mais avançados (com o método OO do matplotlib)

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
# método pyplot (feio)
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# plotar dados com o método orientado a objeto

fig, ax = plt.subplots(dpi=500, figsize=(10, 6))

colors = {'0':'tab:blue', '1':'tab:orange'}

scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# customizar o plot
ax.set(title="Heart disease and cholesterol levels",
       xlabel="Age",
       ylabel="Cholesterol",
       xlim=[45,85])

# adicionar legenda
ax.legend(*scatter.legend_elements(), title="Target");

# adicionar linha horizontal

ax.axhline(y=over_50["chol"].mean(),
           linestyle="--");
```

### Customizing styles

```python
# subplot of chl, age, thalach
plt.style.use('seaborn-white')

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10), dpi=500,
                               sharex=True)

# add data to ax0

scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

# customize ax0

ax0.set(title="Heart disease and cholesterol levels",
        ylabel="Cholesterol")

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# add a meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");



# add data to ax1

scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="Set1")

# customize ax1

ax1.set(title="Heart disease and max heart rate levels",
        xlabel="Age",
        ylabel="Max heart rate")

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# add a meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

# add a title to the figure

fig.suptitle("Heart disease analysis", fontsize=20, fontweight="bold", y=1.05);
```
